```python
import pandas as pd
```

```python
Labor_Data= pd.read_csv('../data/Labour Statistics - Sheet3.csv')
```

```python
Labor_Data
#We have this variable to ensure that we can always access the original dataset. 
```

```python
df= pd.DataFrame(Labor_Data)
#Putting this into a dataframe
```

```python
df = df.sort_values('Location quotient', ascending=False)
#Arranging it according to the location
```

```python
#We try to drop null values here but we realize that our Annual wage is being treated as string.
df['Annual mean wage'] = df[df['Annual mean wage'] > 0]['Annual mean wage']
df.dropna()
```

```python
type(df['Annual mean wage'][0])
#To check that Annual wage is infact a string.
```

```python
df["Annual mean wage"]= df["Annual mean wage"].astype(dtype = float)
#Converting the Annual wage to float/int
#But since we have the $ sign before the numeric it'll throw the error that it can't convert the $91,630. 
```

```python
df['Annual mean wage'] = df['Annual mean wage'].str.replace("$","")
#replacing $ sign in the column with nothing
```

```python
df['Annual mean wage'] = df['Annual mean wage'].str.replace(" ","")
#replacing spaces in the column with nothing
```

```python
df['Annual mean wage'].dropna(inplace = True) 
#Replacing any 0 values in Annual Wage
```

```python
df
#checking how the dataframe looks
```

```python
df = df.sort_values('Annual mean wage', ascending=False)
#arranging it in ascending order
```

```python
df
```

```python
df.to_csv('../data/BLS_Chart.csv', index=False)  
#Saving it as a csv in the data folder.
```
